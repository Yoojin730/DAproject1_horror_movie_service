```python
import requests as req
from bs4 import BeautifulSoup as bs
from tqdm import tqdm_notebook as tn #얼마나 진행됐다 보는 라이브러리

#총 페이지 수 구하기
total_review = soup.select_one('strong.total > em').text

total_review = total_review.replace(',','')

pageNum = int(total_review)//10

if pageNum%10>0:
    pageNum += 1

print('전체 페이지 수:', pageNum)

f = open('movie_review.txt', 'w')

for page in tn(range(pageNum)):
    url = 'https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=164932&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false'

    header = { #user-agent 웹
        'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36',
    }

    res = req.get(url, headers = header, params={'page':page+1})
    soup = bs(res.text, 'html.parser')
    
    #print('-'*10, '페이지번호:', page+1, '-'*10)
    
    reples = soup.select('.score_reple p span')

    for reple in reples:
        if reple.get('class') == None: #get(요소의 속성): 요소의 속성값을 접근하는 함수
                                    #<span id='main'></span>
                                    #get('id') => main
            #print(reple.text.strip())
            f.write(reple.text.strip())
f.close()
print('리뷰데이터 수집 완료!')
```
